```python
import numpy as np
import pandas as pd
proc = pd.read_csv('/content/PP_recipes.csv', sep=',')
raw = pd.read_csv('/content/RAW_recipes.csv', sep=',')
```

```python
merged = pd.merge(proc, raw, left_on='id', right_on='id', how='inner')
#merged
```

```python
merged = merged.drop(columns=['i', 'ingredient_tokens', 'name_tokens', 'steps_tokens', 'techniques', 'contributor_id', 'submitted', 'tags', 'ingredients'])
```

```python
merged['total_fat'] = 0.0
merged['sugar'] = 0.0
merged['sodium'] = 0.0
merged['protein'] = 0.0
merged['saturated_fat'] = 0.0
merged['calories'] = 0.0
merged['carb'] = 0.0
```

```python
str_list = ['calories', 'total_fat', 'sugar', 'sodium', 'protein', 'saturated_fat', 'carb']
for index, row in merged.iterrows():
    arr = row['nutrition'][1:-1].split(',')
    for i in range(len(arr)):
        #calories (#), total fat (PDV), sugar (PDV) , sodium (PDV) , protein (PDV) , saturated fat
        merged[str_list[i]][index] = arr[i]
```

```python
merged = merged.drop(columns=['calorie_level'])
```

```python
user = pd.read_csv('/content/PP_users.csv', sep=',')
```

```python
merged['num_rating'] = 0
merged['ratings'] = 0.0
```

```python
for index, row in user.iterrows():
    items = row['items'][1:-1].split(',')
    ratings = row['ratings'][1:-1].split(',')
    for i in range(len(items)):
        item = int(items[i])
        merged['num_rating'][item] += 1
        merged['ratings'][item] += float(ratings[i])
merged
```

```python
merged = merged.drop(columns=['nutrition'])
```

```python
merged['ratings'] = merged['ratings'] / merged['num_rating']
```

```python
merged['ratings'] = merged['ratings'].replace(np.nan, 0.0)
```

```python
merged
```

```python
data = []
for index, row in merged.iterrows():
    ids = row['ingredient_ids'][1:-1].split(',')
    for i in range(len(ids)):
        data.append([row['id'], int(ids[i])])    
```

```python
df = pd.DataFrame(data, columns=['recipe_id', 'ingr_id'])
df.to_csv('/content/has_ingr.csv')
```

```python
import pandas as pd
with open('/content/ingr.csv', 'r') as csvfile:
    text = csvfile.readlines()
list_names, list_is_meat, list_is_nut = ['']*len(text), [0]*len(text), [0]*len(text)
for i in range(len(text)):
    arr = text[i].split(',')
    name = text[i][len(arr[0])+1:-5]
    index, is_nut, is_meat = int(arr[0]), int(arr[-1]), int(arr[-2])
    list_names[index] = name
    list_is_meat[index] = is_meat
    list_is_nut[index] = is_nut
list_names
```

```python
checklist = ['pork', 'chicken', 'duck', 'turkey', 'lamb', 'beef', 'rabbit', 'poultry',\
             'venison', 'bacon', 'goose', 'pig', 'deer', 'cow', 'liver', 'sausage', 'sausages',\
             'liverwurst', 'sheep', 'fish', 'salmon', 'tuna', 'ham', 'mutton', 'seafood',\
             'salami', 'pheasant', 'boar', 'dog', 'bone', 'bones', 'shrimp', 'tilapia',\
             'fillet', 'fillets', 'cod', 'sardine', 'sardines', 'catfish', 'steak', 'steaks',\
             'mahi', 'halibut', 'trout', 'herring', 'swordfish', 'anchovy', 'anchovies',\
             'bass', 'hake', 'flounder', 'monkfish', 'tenderloin', 'tenderloins', 'lobster', 'lobsters',\
             'meat', 'roe', 'clam', 'clams', 'lard', 'rib', 'ribs', 'roast', 'brisket', 'veal',\
             'filet mignon']

for i in range(len(list_names)):
    words = list_names[i].split(' ')
    if 'vegetarian' in words or 'imitation' in words:
        continue
    for j in range(len(checklist)):
        if checklist[j] in words:
            list_is_meat[i] = 1
            break
```

```python
import numpy as np
arr = np.load('/content/ingr_map.pkl', allow_pickle=True)
ids = list(arr['id'])
ingrs = list(arr['raw_ingr'])
words = list(arr['raw_words'])
processed = list(arr['processed'])
len_proc = list(arr['len_proc'])
replaced = list(arr['replaced'])
count = list(arr['count'])
set_ingrs = set(ingrs)

ingr_map = {}
for i in range(len(ids)):
    if ids[i] not in ingr_map:
        ingr_map[ids[i]] = replaced[i] + ',' + str(list_is_meat[i]) + ',' + str(list_is_nut[i])
    else:
        tmp_is_meat, tmp_is_nut = int(ingr_map[ids[i]].split(',')[1]), int(ingr_map[ids[i]].split(',')[2])
        tmp_is_meat, tmp_is_nut = list_is_meat[i] and tmp_is_meat, list_is_nut[i] and tmp_is_nut
        ingr_map[ids[i]] = replaced[i] + ',' + str(tmp_is_meat) + ',' + str(tmp_is_nut)
print(len(ingr_map))
```

```python
out_text = 'id,name,is_meat,is_nut\n'
for id in ingr_map.keys():
    new_line = str(id) + ',' + ingr_map[id] + '\n'
    out_text += new_line
f = open('/content/ingr_map.csv', 'w')
f.write(out_text)
f.close()
```
